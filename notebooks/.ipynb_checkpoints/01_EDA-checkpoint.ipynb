```python
# Imports et chargement
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Affichage inline pour Jupyter
%matplotlib inline

# Charge le dataset
df = pd.read_csv("../data/dataset_risque_defaut.csv")
df.head()
```

```python
# Aperçu
print("Shape:", df.shape)
print("\nTypes :")
print(df.dtypes)
print("\nValeurs nulles par colonne :")
print(df.isna().sum())
```

```python
# Statistiques numériques
df.describe(percentiles=[0.01,0.05,0.25,0.5,0.75,0.95,0.99]).T
```

```python
# Cible
print(df['defaut'].value_counts())
print(df['defaut'].value_counts(normalize=True))

sns.countplot(x='defaut', data=df)
plt.title("Distribution variable cible (defaut)")
plt.show()
```

```python
# Histogrammes
cols = ['age', 'revenu_mensuel', 'montant_credit', 'duree_mois', 'taux_endettement', 'ratio_dette_revenu']
df[cols].hist(bins=30, figsize=(12,8))
plt.suptitle("Histogrammes variables numériques")
plt.show()
```

```python
# Boxplots
plt.figure(figsize=(12,6))
sns.boxplot(data=df[['revenu_mensuel','montant_credit','ratio_dette_revenu']])
plt.title("Boxplots - repérage outliers")
plt.show()
```

```python
# Corrélation
plt.figure(figsize=(10,8))
corr = df.select_dtypes(include=[np.number]).corr()
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Matrice de corrélation (numériques)")
plt.show()
```

```python
# Retards vs defaut
retard_cols = ['nb_retards_30j','nb_retards_60j','nb_retards_90j']
for c in retard_cols:
    plt.figure(figsize=(6,4))
    sns.boxplot(x='defaut', y=c, data=df)
    plt.title(f"{c} par defaut")
    plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='defaut', y='ratio_dette_revenu', data=df)
plt.title("ratio_dette_revenu par defaut")
plt.show()
```

```python
# Synthèse automatique (point de départ)
problems = {}
# Vérifier missing
missing = df.isna().sum()
problems['missing_values'] = missing[missing>0].to_dict()
# Vérifier valeurs extrêmes (ex: >99e percentile)
outliers = {}
for c in df.select_dtypes(include=[np.number]).columns:
    p99 = df[c].quantile(0.99)
    p01 = df[c].quantile(0.01)
    outliers[c] = {'p01': p01, 'p99': p99}
problems['outliers_01_99'] = outliers

import json
print("Problèmes détectés (extrait) :")
print(json.dumps({k: v for k,v in problems.items() if v}, indent=2)[:1000])
```
